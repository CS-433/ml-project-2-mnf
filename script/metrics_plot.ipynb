```python
import numpy as np
import os

import ast

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
data_unet = pd.read_csv('./training_metrics_model_UNET_100.csv')
data_maresunet = pd.read_csv('./training_metrics_model_MARESUNET_100.csv')
data_resnet = pd.read_csv('./training_metrics_model_RESNET_100.csv')
```

```python
data_unet.shape
```

```python
data_maresunet.shape
```

```python
data_resnet.shape
```

```python
data_unet['Train Loss'] = data_unet['Train Loss'].apply(lambda x: ast.literal_eval(x)[1])
data_unet['Train Accuracy'] = data_unet['Train Accuracy'].apply(lambda x: ast.literal_eval(x)[1])
data_unet['Train F1'] = data_unet['Train F1'].apply(lambda x: ast.literal_eval(x)[1])

data_unet.head()
```

```python
data_maresunet['Train Loss'] = data_maresunet['Train Loss'].apply(lambda x: ast.literal_eval(x)[1])
data_maresunet['Train Accuracy'] = data_maresunet['Train Accuracy'].apply(lambda x: ast.literal_eval(x)[1])
data_maresunet['Train F1'] = data_maresunet['Train F1'].apply(lambda x: ast.literal_eval(x)[1])

data_maresunet.head()
```

```python
data_resnet['Train Loss'] = data_resnet['Train Loss'].apply(lambda x: ast.literal_eval(x)[1])
data_resnet['Train Accuracy'] = data_resnet['Train Accuracy'].apply(lambda x: ast.literal_eval(x)[1])
data_resnet['Train F1'] = data_resnet['Train F1'].apply(lambda x: ast.literal_eval(x)[1])

data_resnet.head()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(13, 5))

sns.lineplot(data=data_unet, x='Epoch', y='Train Loss', ax=axes[0], color = '#ecad3c', label='UNet')
sns.lineplot(data=data_resnet, x='Epoch', y='Train Loss', ax=axes[0], color = '#c0283c', label='ResNet')
sns.lineplot(data=data_maresunet, x='Epoch', y='Train Loss', ax=axes[0], color = '#143786', label='MAResUNet')
axes[0].set_title('Train Loss')

sns.lineplot(data=data_unet, x='Epoch', y='Train Accuracy', ax=axes[1], color = '#ecad3c', label='UNet')
sns.lineplot(data=data_resnet, x='Epoch', y='Train Accuracy', ax=axes[1], color = '#c0283c', label='ResNet')
sns.lineplot(data=data_maresunet, x='Epoch', y='Train Accuracy', ax=axes[1], color = '#143786', label='MAResUNet')
axes[1].set_title('Train Accuracy')

sns.lineplot(data=data_unet, x='Epoch', y='Train F1', ax=axes[2], color = '#ecad3c', label='UNet')
sns.lineplot(data=data_resnet, x='Epoch', y='Train F1', ax=axes[2], color = '#c0283c', label='ResNet')
sns.lineplot(data=data_maresunet, x='Epoch', y='Train F1', ax=axes[2], color = '#143786', label='MAResUNet')
axes[2].set_title('Train F1')

plt.tight_layout()
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(13, 5))

sns.lineplot(data=data_unet, x='Epoch', y='Val Loss', ax=axes[0], color = '#ecad3c', label='UNet')
sns.lineplot(data=data_resnet, x='Epoch', y='Val Loss', ax=axes[0], color = '#c0283c', label='ResNet')
sns.lineplot(data=data_maresunet, x='Epoch', y='Val Loss', ax=axes[0], color = '#143786', label='MAResUNet')

sns.lineplot(data=data_unet, x='Epoch', y='Val Accuracy', ax=axes[1], color = '#ecad3c', label='UNet')
sns.lineplot(data=data_resnet, x='Epoch', y='Val Accuracy', ax=axes[1], color = '#c0283c', label='ResNet')
sns.lineplot(data=data_maresunet, x='Epoch', y='Val Accuracy', ax=axes[1], color = '#143786', label='MAResUNet')

sns.lineplot(data=data_unet, x='Epoch', y='Val F1', ax=axes[2], color = '#ecad3c', label='UNet')
sns.lineplot(data=data_resnet, x='Epoch', y='Val F1', ax=axes[2], color = '#c0283c', label='ResNet')
sns.lineplot(data=data_maresunet, x='Epoch', y='Val F1', ax=axes[2], color = '#143786', label='MAResUNet')

for ax in axes:
    ax.set_yscale('log')

axes[0].set_title('Validation Loss (Log Scale)')
axes[0].set_ylabel('Loss (Log Scale)')
axes[1].set_title('Validation Accuracy (Log Scale)')
axes[1].set_ylabel('Accuracy (Log Scale)')
axes[2].set_title('Validation F1 (Log Scale)')
axes[2].set_ylabel('F1 (Log Scale)')

plt.tight_layout()
plt.legend()
plt.show()
```

```python
models = ['UNET', 'ResNET', 'MAResUNET']
f1_scores = [0.870, 0.877, 0.895]
accuracies = [0.929, 0.932, 0.949]
runtimes_minutes = [38 + 13/60, 46 + 5/60, 10 + 12/60]

fig, axs = plt.subplots(1, 3, figsize=(10, 3))

# F1-score subplot
axs[0].bar(models, f1_scores, color=['#ecad3c', '#c0283c', '#143786'])
axs[0].set_ylabel('F1-score')
axs[0].set_title('F1-score')
axs[0].set_ylim(min(f1_scores) - 0.02, max(f1_scores) + 0.02)

# Accuracy subplot
axs[1].bar(models, accuracies, color=['#ecad3c', '#c0283c', '#143786'])
axs[1].set_ylabel('Accuracy')
axs[1].set_title('Accuracy')
axs[1].set_ylim(min(accuracies) - 0.01, max(accuracies) + 0.01)

# Runtime subplot
axs[2].bar(models, runtimes_minutes, color=['#ecad3c', '#c0283c', '#143786'])
axs[2].set_ylabel('Runtime (min)')
axs[2].set_title('Runtime')
axs[2].set_ylim(min(runtimes_minutes) - 5, max(runtimes_minutes) + 5)

plt.tight_layout()
plt.show()
```
